**2.2. Cleaning the data**

You might find you need to transpose your dataframe, check out what its row indexes are, and reset the index. You  also might find you need to assign the values of the first row to your column headings  . (Hint: recall the .columns feature of DataFrames, as well as the iloc[] method).

```python
# transpose dataframe and reset index
properties_t = properties.transpose()
properties_t = properties_t.reset_index()

# assign first row to a variable, new_header
# keep all remaining rows in the dataframe
# assign new_header to column headings
new_header = properties_t.iloc[0] 
properties_t = properties_t[1:] 
properties_t.columns = new_header 

# display the shape and header of the dataframe
print("Rows, columns: " + str(properties_t.shape))
properties_t.head()
```

**2.3. Cleaning the data (part 2)**

You might we have to **rename** a couple columns. How do you do this? The clue's pretty bold...

```python
# rename the first to columns as 'Borough' and 'GSS Code'
properties_t = properties_t.rename(columns={'Unnamed: 0': 'Borough', pd.NaT: 'GSS Code'})
properties_t.head()
```

**2.4.Transforming the data**

Remember what Wes McKinney said about tidy data? 

You might need to **melt** your DataFrame here. 

```python
# melt data frame
clean_props = pd.melt(properties_t, id_vars=['Borough', 'GSS Code'])
clean_props.head()
```

```python
# rename columns to 'Month' and 'Avg Price'
clean_props = clean_props.rename(columns={0: 'Month', 'value': 'Avg Price'})
clean_props.head()
```

```python
# change 'Avg Price' to float, check datatypes again
clean_props['Avg Price'] = pd.to_numeric(clean_props['Avg Price'])
clean_props.dtypes
```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
# rename dataframe to match instructions
dfg = df_years_prices

def create_price_ratio(dfg_filtered):
    '''take a filter of the dataframe for some borough'''
    '''get the Avg Price for the years 1998 and 2000'''
    '''calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018'''
    '''return the ratio'''
    avg1998 = float(dfg_filtered['Avg Price'][dfg_filtered['Year']==1998])
    avg2018 = float(dfg_filtered['Avg Price'][dfg_filtered['Year']==2018])
    ratio = [avg1998/avg2018]
    return ratio
```

```python
create_price_ratio(dfg[dfg['Borough']=='Hammersmith & Fulham'])
```

```python
# create an empty dictionary that will be fill with boroughs and ratios
ratios = {}

# iterate through each of the unique elements of the Borough column of dfg
# add key, Borough, and value, ratio, to the dictionary
for i in dfg['Borough'].unique():
    borough = dfg[dfg['Borough'] == i]
    ratios[i] = create_price_ratio(borough)

ratios
```

```python
# create a dataframe from the ratios dictionary
# rename columns as Borough and Ratio
df_ratios = pd.DataFrame(ratios).transpose().reset_index()
df_ratios.rename(columns = {'index':'Borough', 0:'Ratio'}, inplace=True)
df_ratios
```

```python
# sort ratios to see which is largest
df_sorted = df_ratios.sort_values(by='Ratio',ascending=False)
df_sorted
```

```python
# use the head of the dataframe for plotting largest ratios
df_plot = df_sorted.head(10)
# plot ratios with y axis inverted so largest appears on top
fig, ax = plt.subplots()
ax.barh(df_plot['Borough'], df_plot['Ratio'])
ax.invert_yaxis()
```
